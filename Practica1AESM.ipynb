```python
class Persona:

    lista=[]
    
    def __init__(self,nombre,correo):
        self.nombre=nombre
        self.correo=correo

    def registrar(self):
        Persona.lista.append(self)
        print(f"La persona {self.nombre} ha sido registrada con el correo {self.correo}")

    def actualizar_datos(self,nombre,correo):
        self.nombre=nombre
        self.correo=correo
        print(f"Los datos han sido actualizados")

    @classmethod
    def personas_registradas(cls):
        print("Personas registradas")
        for Persona in cls.lista:
            print(f"-{Persona.nombre} - {Persona.correo}")
```

```python
class Usuario(Persona):
    def __init__(self, nombre, correo):
        super().__init__(nombre, correo)
        self.historial_reservas = []

    def reservar(self, funcion, asientos):
        if asientos <= funcion.asientos_disponibles:
            funcion.asientos_disponibles -= asientos
            self.historial_reservas.append({"funcion": funcion, "asientos": asientos})
            print(f"Reserva realizada para '{funcion.pelicula.titulo}' en la sala {funcion.sala.identificador}.")
        else:
            print("No hay suficientes asientos disponibles.")

    def cancelar_reserva(self, funcion):
        reserva = next((r for r in self.historial_reservas if r["funcion"] == funcion), None)
        if reserva:
            funcion.asientos_disponibles += reserva["asientos"]
            self.historial_reservas.remove(reserva)
            print(f"Reserva cancelada para '{funcion.pelicula.titulo}'.")
        else:
            print("No tienes una reserva para esta función.")
```

```python
class Empleado(Persona):
    def __init__(self, nombre, correo, rol):
        super().__init__(nombre, correo)
        self.rol = rol

    def agregar_funcion(self, funcion):
        print(f"Función agregada: {funcion.pelicula.titulo} a las {funcion.hora} en la sala {funcion.sala.identificador}.")

    def modificar_promocion(self, promocion, nuevo_descuento, nuevas_condiciones):
        promocion.descuento = nuevo_descuento
        promocion.condiciones = nuevas_condiciones
        print(f"Promoción modificada: {nuevo_descuento}% de descuento. {nuevas_condiciones}.")
```

```python
class Espacio:
    def __init__(self,capacidad,identificador):
        self.capacidad=capacidad
        self.identificador=identificador
    
    def descripcion(self):
        print(f"El edificio tiene tamaño {self.capacidad} y tiene id {self.identificador}")
```

```python
class Sala(Espacio):
    def __init__(self,capacidad,identificador,tipo):
        super().__init__(capacidad,identificador)
        self.tipo=tipo
        self.disponibilidad=True

    def Consultardisponibilidad(self):
        if self.disponibilidad:
            print("La sala esta disponible")
        else:
            print("La sala esta ocupada")
```

```python
class Pelicula:
    def __init__(self, titulo, genero, duracion):
        self.titulo = titulo
        self.genero = genero
        self.duracion = duracion
```

```python
class Funcion:
    def __init__(self, pelicula, sala, hora, asientos_disponibles=None):
        self.pelicula = pelicula
        self.sala = sala
        self.hora = hora
        self.asientos_disponibles = asientos_disponibles or sala.capacidad
```

```python
class Promocion:
    def __init__(self, descuento, condiciones):
        self.descuento = descuento
        self.condiciones = condiciones

    def mostrar(self):
        print(f"Promoción: {self.descuento}% de descuento. Condiciones: {self.condiciones}")

```

```python

class Persona:
    def __init__(self, nombre, correo):
        self.nombre = nombre
        self.correo = correo


class Empleado(Persona):
   
    def __init__(self, nombre, correo, rol):
        super().__init__(nombre, correo)
        self.rol = rol

    def tomar_pedido(self, pedido):
        pedido.estado = "en preparación"
        print(f"{self.nombre} ({self.rol}) está procesando el pedido de {pedido.cliente.nombre}...")

    def consultar_inventario(self, inventario):
        
        print(f"\n{self.nombre} está consultando el inventario:")
        for ing, cant in inventario.ingredientes.items():
            print(f" - {ing}: {cant} unidades")


class ProductoBase:
   
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def __str__(self):
        return f"{self.nombre} - ${self.precio}"


class Bebida(ProductoBase):
    
    def __init__(self, nombre, precio, tamano, temperatura, endulzante):
        super().__init__(nombre, precio)
        self.tamano = tamano         
        self.temperatura = temperatura  
        self.endulzante = endulzante  

    def personalizar(self, tamano=None, temperatura=None, endulzante=None):
       
        if tamano:
            self.tamano = tamano
        if temperatura:
            self.temperatura = temperatura
        if endulzante:
            self.endulzante = endulzante

    def __str__(self):
        return (f"{self.nombre} ({self.tamano}, {self.temperatura}, {self.endulzante}) "
                f"- ${self.precio}")


class Comida(ProductoBase):
   
    def __init__(self, nombre, precio, es_vegetariano=False):
        super().__init__(nombre, precio)
        self.es_vegetariano = es_vegetariano

    def __str__(self):
        tipo = "Vegetariano" if self.es_vegetariano else "No vegetariano"
        return f"{self.nombre} ({tipo}) - ${self.precio}"




class Inventario:
   
    def __init__(self):
 
        self.ingredientes = {}

    def agregar_ingrediente(self, ingrediente, cantidad):
       
        if ingrediente in self.ingredientes:
            self.ingredientes[ingrediente] += cantidad
        else:
            self.ingredientes[ingrediente] = cantidad
        print(f"Se agregaron {cantidad} unidades de '{ingrediente}' al inventario.")

    def verificar_disponibilidad(self, ingrediente, cantidad):
        
        return self.ingredientes.get(ingrediente, 0) >= cantidad

    def usar_ingrediente(self, ingrediente, cantidad):
       
        if self.verificar_disponibilidad(ingrediente, cantidad):
            self.ingredientes[ingrediente] -= cantidad
            print(f"Se usaron {cantidad} unidades de '{ingrediente}'.")
            return True
        else:
            print(f"No hay suficiente '{ingrediente}' en el inventario.")
            return False




class Promocion:
    
    def __init__(self, nombre, descuento):
      
        self.nombre = nombre
        self.descuento = descuento

    def aplicar(self, pedido):
       
        total_original = pedido.calcular_total()
        monto_descuento = total_original * (self.descuento / 100)
        nuevo_total = total_original - monto_descuento
        print(f"Promoción '{self.nombre}' aplicada. Descuento: ${monto_descuento:.2f}")
        return nuevo_total




class Pedido:
  
    def __init__(self, cliente):
        self.cliente = cliente
        self.productos = []
        self.estado = "pendiente"
        self.promocion_aplicada = None

    def agregar_producto(self, producto):
        self.productos.append(producto)
        print(f"Se agregó al pedido: {producto}")

    def eliminar_producto(self, producto):
        if producto in self.productos:
            self.productos.remove(producto)
            print(f"Se eliminó del pedido: {producto}")
        else:
            print(f"El producto {producto.nombre} no está en el pedido.")

    def aplicar_promocion(self, promocion):
        self.promocion_aplicada = promocion

    def calcular_total(self):
        return sum([prod.precio for prod in self.productos])

    def mostrar_detalles(self):
        print(f"\n--- Detalles del pedido de {self.cliente.nombre} ---")
        for prod in self.productos:
            print(f" - {prod}")
        print(f"Estado: {self.estado}")
        total = self.calcular_total()
        if self.promocion_aplicada:
            total = self.promocion_aplicada.aplicar(self)
        print(f"Total a pagar: ${total:.2f}")




#
if __name__ == "__main__":

 
    inventario = Inventario()
    inventario.agregar_ingrediente("café", 10)
    inventario.agregar_ingrediente("leche", 5)
    inventario.agregar_ingrediente("azúcar", 20)
    inventario.agregar_ingrediente("pan", 10)
    inventario.agregar_ingrediente("queso", 5)
    inventario.agregar_ingrediente("tomate", 8)

  
    barista = Empleado("Carlos", "carlos@cafeteria.com", "Barista")

   
    cliente1 = Persona("Ana", "ana@example.com")

    pedido1 = Pedido(cliente1)

   
    cafe_americano = Bebida(nombre="Café Americano", precio=30, tamano="mediano", 
                            temperatura="caliente", endulzante="azúcar")

 
    cafe_americano.personalizar(endulzante="sin azúcar")

    sandwich_queso = Comida(nombre="Sándwich de Queso", precio=50, es_vegetariano=True)


    pedido1.agregar_producto(cafe_americano)
    pedido1.agregar_producto(sandwich_queso)


    barista.tomar_pedido(pedido1)

    promo_cliente_frecuente = Promocion(nombre="Cliente Frecuente", descuento=10)
    pedido1.aplicar_promocion(promo_cliente_frecuente)

    pedido1.mostrar_detalles()

    barista.consultar_inventario(inventario)

   
    inventario.usar_ingrediente("café", 1)
    inventario.usar_ingrediente("leche", 1)

    inventario.usar_ingrediente("pan", 1)
    inventario.usar_ingrediente("queso", 1)
    inventario.usar_ingrediente("tomate", 1)

  
    pedido1.estado = "finalizado"
    pedido1.mostrar_detalles()
```

BIBLIOTECA


```python
class Persona:
    lista = []

    def __init__(self, nombre, correo):
        self.nombre = nombre
        self.correo = correo

    def registrar(self):
        Persona.lista.append(self)
        print(f"La persona {self.nombre} ha sido registrada con el correo {self.correo}")


class Usuario(Persona):
    def __init__(self, nombre, correo):
        super().__init__(nombre, correo)
        self.materiales_prestados = []
        self.multa = 0  

    def tomar_prestado(self, material):
        if material.disponible:
            self.materiales_prestados.append(material)
            material.disponible = False
            print(f"{self.nombre} ha tomado prestado: {material.titulo}")
        else:
            print(f"{material.titulo} no está disponible")

    def devolver_material(self, material, dias_retraso=0):
        if material in self.materiales_prestados:
            self.materiales_prestados.remove(material)
            material.disponible = True
            if dias_retraso > 0:
                multa = dias_retraso * 1  
                self.multa += multa
                print(f"{self.nombre} devolvió {material.titulo} con {dias_retraso} días de retraso. Se le aplica una multa de ${multa}. Total acumulado: ${self.multa}")
            else:
                print(f"{self.nombre} devolvió {material.titulo} a tiempo.")
        else:
            print(f"{self.nombre} no tiene {material.titulo} prestado.")


class Bibliotecario(Persona):
    def __init__(self, nombre, correo):
        super().__init__(nombre, correo)

    def agregar_material(self, sucursal, material):
        sucursal.agregar_material(material)
        print(f"El bibliotecario {self.nombre} agregó {material.titulo} a la sucursal {sucursal.nombre}.")

    def transferir_material(self, origen, destino, material):
        if material in origen.catalogo:
            origen.remover_material(material)
            destino.agregar_material(material)
            print(f"El bibliotecario {self.nombre} transfirió {material.titulo} de {origen.nombre} a {destino.nombre}.")
        else:
            print(f"{material.titulo} no se encuentra en la sucursal {origen.nombre}.")


class Material:
    def __init__(self, titulo, tipo):
        self.titulo = titulo
        self.tipo = tipo
        self.disponible = True


class Sucursal:
    def __init__(self, nombre):
        self.nombre = nombre
        self.catalogo = []  

    def agregar_material(self, material):
        self.catalogo.append(material)
        print(f"Se agregó {material.titulo} a la sucursal {self.nombre}.")

    def remover_material(self, material):
        if material in self.catalogo:
            self.catalogo.remove(material)
            print(f"Se removió {material.titulo} de la sucursal {self.nombre}.")
        else:
            print(f"{material.titulo} no se encuentra en la sucursal {self.nombre}.")


def buscar_materiales_disponibles(sucursales):

    disponibles = []
    for sucursal in sucursales:
        for material in sucursal.catalogo:
            if material.disponible:
                disponibles.append((sucursal.nombre, material))
    return disponibles


sucursal1 = Sucursal("Central")
sucursal2 = Sucursal("Norte")


libro1 = Material("2005", "Libro")
revista1 = Material("Crepúsculo", "Revista")
libro2 = Material("Python Avanzado", "Libro")


bibliotecario1 = Bibliotecario("Sofía", "sofiacalvariaa@email.com")
usuario1 = Usuario("Ximena", "ximena06@email.com")
usuario1.registrar()

bibliotecario1.agregar_material(sucursal1, libro1)
bibliotecario1.agregar_material(sucursal1, revista1)
bibliotecario1.agregar_material(sucursal2, libro2)

sucursales = [sucursal1, sucursal2]
disponibles = buscar_materiales_disponibles(sucursales)
print("\nMateriales disponibles:")
for nombre_sucursal, material in disponibles:
    print(f"Sucursal: {nombre_sucursal} - {material.titulo} ({material.tipo})")


usuario1.tomar_prestado(libro1)

disponibles = buscar_materiales_disponibles(sucursales)
print("\nMateriales disponibles después del préstamo:")
for nombre_sucursal, material in disponibles:
    print(f"Sucursal: {nombre_sucursal} - {material.titulo} ({material.tipo})")


usuario1.devolver_material(libro1, dias_retraso=3)

bibliotecario1.transferir_material(sucursal1, sucursal2, revista1)

print("\nCatálogo en sucursales después de la transferencia:")
for sucursal in sucursales:
    print(f"Sucursal {sucursal.nombre}: {[material.titulo for material in sucursal.catalogo]}")
```
